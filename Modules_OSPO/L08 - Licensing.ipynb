## Types of Software Licenses

Understanding software licensing is crucial for navigating the open source landscape. Here are the four main categories:

### 1. Proprietary Software
- **Characteristics:** Closed source, proprietary rights reserved by the owner
- **Examples:** Microsoft Windows, Adobe Photoshop, many commercial applications
- **User Rights:** Very limited; typically just the right to use

### 2. Source Available
- **Characteristics:** Source code is visible, but comes with restrictions (e.g., no modifications or redistribution)
- **Examples:** Sentry, OctoML, Meta's Llama license(s), other specific enterprise software
- **User Rights:** Can view source but limited modification/redistribution rights

### 3. Permissive Open Source
- **Characteristics:** Minimal restrictions, allows proprietary derivatives
- **Examples:** MIT/BSD License, Apache 2.0, Mozilla Public License
- **User Rights:** Can use, modify, and redistribute, including in proprietary projects - most permissive licenses (but not all) respect the 4 freedoms of free software

### 4. Copyleft Open Source
- **Characteristics:** Ensures derivatives remain free; "viral" licensing
- **Examples:** GPL (General Public License), LGPL, AGPL
- **User Rights:** All four freedoms guaranteed, with the requirement that derivatives also be free - this distinction is also called "strong copyleft" and is the basis of the Free Software Foundation's definition of free software.


> **Note:** The Free Software Foundation considers only copyleft licenses that guarantee all four freedoms to be "free software." Many permissive licenses are considered "free software" by the Free Software Foundation, but some permissive licenses (e.g., the [Open Watcom License](https://www.gnu.org/philosophy/open-source-misses-the-point.en.html#:~:text=First,such%20licenses)) do not meet the criteria for free software.

> **Important!** Free software and Open-Source can be used interchangeably, but they have different philosophical underpinnings. All licenses shown past level 3 (permissive open source) are considered "free software" by the Free Software Foundation, but only copyleft licenses (level 4) can guarantee all four freedoms.

#### GPL-Compatible Licenses

Any license that meets the criteria for free software (i.e., guarantees all four freedoms) is considered "GPL-compatible." This includes popular licenses such as:
- **Apache 2.0**
- **LGPLv3**: Lesser General Public License
- **GPLv3**: General Public License
- **MIT License**: Also known as the Expat License
- **BSD Licenses**: BSD 2-Clause, BSD 3-Clause, etc. (note that BSD 0-Clause is completely separate from the other BSD licenses)

> For a comprehensive list of free software licenses and their compatibility status, ***see the [FSF's license list](https://www.gnu.org/licenses/license-list.html)***, which provides detailed analysis of virtually all modern FLOSS software licenses.

## *Open Source* Licenses

Intellectual property primarily consists of *copyrights*, *patents*, and *trademarks*. Software licensing primarily focuses on *copyright* as it pertains to code use and attribution. The lack of any license by default implies that the code is under exclusive copyright *separately* for each individual contributor. Collaborative work is difficult under these conditions. Licenses explicitly grant or deny permissions for contributors in different ways that suit the needs of various projects. In this section, only *open source* licenses (permissive and copyleft) will be discussed. Nonetheless, it should be noted that software can be licensed under both *open source* and *proprietary* licenses as is the case with MySQL (Community Edition - GPLv2 vs. Enterprise Edition).

Selecting the right license is crucial for your open source project! It is suggested to briefly skim each license. A synopsis of granted rights for all of them can be found on GitHub's `LICENSE` creation tool or at [ChooseALicense.com](https://choosealicense.com/). *Generally*, copyleft licensing requires strictly requires software utilizing copyleft software to use the same copyleft license. Thus, these licenses are usually incompatible with all other licenses unless explicitly stated otherwise. Permissive licensing ensures some attribution to original software (unlike public domain) without major restrictions on the software itself. A consequence of this is that closed source/proprietary software can utilize permissively licensed open source software. Commercial companies may be more willing to contribute and use permissively licensed software, though this should not imply that corporations are unwilling to contribute or use copyleft software. Thus, it is important to consider the target audience of your software and the potential contributors to that software.

Most OSS licenses allow for *commercial use*, *distribution*, *modification*, and *private use*. Most OSS licenses also include a limitation of liability ("the software is provided on an 'as-is' basis") and an explicit statement of *no warranty*. All GPL-family licenses (and some others) explicitly require documentation of all changes to the codebase. Additionally, all GPL-family licenses require that source code be made available with the distribution of licensed material.

Since copyleft licensing typically requires the ***same*** license to be used in derivative works, the license is inherently viral. However, custom exceptions can be made to allow for relicensing to a different licenses regardless of the original license. Furthermore, if all copyright holders (generally just the contributors) agree to relicense, then all future software may be relicensed under a new license. License changes do not retroactively affect previous versions unless there is an explicit clause allowing for the use of future versions of a specific license (e.g. "GPLv2 or any later version"). Note that different versions of a license with the same name count as different licenses.

*Most* licenses require a copy of the license and a copyright notice from the source project, usually included in a single file called `LICENSE` located in the root directory of a project. Written attribution to the authors is typically also required in one of the documents.

Some licenses like the Affero General Public License (AGPLv3) specifically define software distribution as including software use over a network. This definition exists to counteract circumvention of GPL licensing via SaaS (Software as a Service).

Other licenses like the Lesser General Public License (LGPLv3/LGPLv2.1) weaken copyleft to allow for dynamic linking of LGPL libraries in non-OSS software. In such a case, there are obligations like providing library source code (like all other GPL licenses) and allowing for the replacement of library with a modified version (typically .so or .dll files). LGPLv3 requires that all information necessary to run modified versions of the LGPL library be made accessible in user-facing products to prevent a process called *tivoization*.

Licenses like the Apache 2.0 license provide explicit patent grant rights to the contributors. Other licenses may explictily deny patent grant rights or in many cases will not mention patent rights at all. In such cases, read each license individually as some contain implicit grants. Some licenses strengthen the patent grant by granting weak retaliation, ensures that the user initiating a patent lawsuit only loses the license for the duration of the case.

Always remember, you should choose a license that aligns with your project's goals and community values.

## Non-Software Licensing

If your OSS project contains significant non-software components (e.g. text, images, videos), then it is important to use one of the licenses under Creative Commons. The `CC0` license effectively puts work into public domain, though it may be better to use a different license in such a scenario. The `CC BY` license (also called `CC BY 4.0`) requires attribution to the authors and a link to the `CC BY` license. There are three modifiers that may be placed on a `CC BY` license:

- SA (ShareAlike): This is similar to copyleft software licensing in that the same license must be passed on
- NC (NonCommercial): The work may not be used to make money in the form of advertisements, products, etc.
- ND (NoDerivatives): The work may not be changed in any way if used

These three modifiers can be combined into various combinations such as `CC BY-ND`, `CC BY-NC-ND`, or `CC BY-NC-SA` (other combinations not listed). Each license variant may also have distinct versions (e.g. 3.0, 4.0).

Remember that just like OSS licenses, Creative Commons deals with open or partially available content and is not proprietary or copyright proof. Besides work done in the federal government, creative work is copyrighted by default.
